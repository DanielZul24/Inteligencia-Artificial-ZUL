Un taller automotriz necesita llevar un control de su inventario de repuestos. Crea un programa que permita agregar, eliminar y consultar la cantidad de repuestos disponibles, así como realizar una compra cuando el inventario de un repuesto está bajo.

```python
# Control de inventario de repuestos
# La clave es el nombre del repuesto y el valor es la cantidad disponible
"""aqui se guardaran los repuestos """
inventario = {}


# Función para agregar un repuesto al inventario
def agregar_repuesto(nombre, cantidad):
    """Agrega un repuesto al inventario.
    Si el repuesto ya existe, se suma la cantidad a la existente.
    Si no existe, se añade como un nuevo repuesto


    nombre - El nombre del repuesto.
    cantidad - La cantidad de repuestos que se va a agregar """
    if nombre in inventario:
        # Si el repuesto ya existe, se suma la cantidad
        inventario[nombre] += cantidad
    else:
        # Si no existe, se crea una nueva entrada en el inventario
        inventario[nombre] = cantidad
    print(f"Se agregaron {cantidad} unidades de {nombre} al inventario.\n")

# Función para eliminar un repuesto del inventario
def eliminar_repuesto(nombre):

    if nombre in inventario:
        # Si el repuesto existe, se elimina del inventario
        del inventario[nombre]
        print(f"El repuesto {nombre} ha sido eliminado del inventario.\n")
    else:
        # Si el repuesto no existe, se muestra un mensaje
        print(f"El repuesto {nombre} no se encuentra en el inventario.\n")

# Función para consultar la cantidad de un repuesto
def consultar_repuesto(nombre):

    if nombre in inventario:
        # Si el repuesto existe, se muestra la cantidad disponible
        print(f"Hay {inventario[nombre]} unidades de {nombre} en el inventario.\n")
    else:
        # Si el repuesto no existe, se muestra un mensaje
        print(f"El repuesto {nombre} no se encuentra en el inventario.\n")

# Función para realizar una compra si el inventario está bajo
def comprar_repuesto(nombre, cantidad_minima, cantidad_a_comprar):

    if nombre in inventario:
        # Verifica si la cantidad es menor a la mínima permitida
        if inventario[nombre] < cantidad_minima:
            print(f"El inventario de {nombre} es bajo. Comprando {cantidad_a_comprar} unidades...\n")
            # Si es así, se suma la cantidad comprada al inventario
            inventario[nombre] += cantidad_a_comprar
        else:
            # Si la cantidad es suficiente, no se hace la compra
            print(f"El inventario de {nombre} es suficiente. No es necesario comprar más.\n")
    else:
        # Si el repuesto no existe, muestra un mensaje pidiendo que lo agreguen primero
        print(f"El repuesto {nombre} no está registrado. Por favor, agréguelo primero.\n")

# En el siguiente menu el usuario podra seleccionar una opcion
def menu():
    """Función principal que muestra un menú interactivo para gestionar el inventario.
    Permite al usuario agregar, eliminar, consultar y comprar repuestos"""
    while True:
        # Muestra las opciones del menú al usuario
        print("=== Taller Automotriz: Gestión de Inventario ===")
        print("1. Agregar repuesto")
        print("2. Eliminar repuesto")
        print("3. Consultar repuesto")
        print("4. Comprar repuesto")
        print("5. Salir")

        # Solicita al usuario que seleccione una opción
        opcion = input("Seleccione una opción: ")

        if opcion == '1':
            # Agregar repuesto
            nombre = input("Nombre del repuesto: ")
            cantidad = int(input("Cantidad a agregar: "))
            agregar_repuesto(nombre, cantidad)

        elif opcion == '2':
            # Eliminar repuesto
            nombre = input("Nombre del repuesto a eliminar: ")
            eliminar_repuesto(nombre)

        elif opcion == '3':
            # Consultar repuesto
            nombre = input("Nombre del repuesto a consultar: ")
            consultar_repuesto(nombre)

        elif opcion == '4':
            # Comprar repuesto si la cantidad es insuficiente
            nombre = input("Nombre del repuesto: ")
            cantidad_minima = int(input("Cantidad mínima permitida: "))
            cantidad_a_comprar = int(input("Cantidad a comprar: "))
            comprar_repuesto(nombre, cantidad_minima, cantidad_a_comprar)

        elif opcion == '5':
            # Salir del programa
            print("Saliendo...")
            break
        else:
            # Manejo de opción no válida
            print("Opción no válida, seleccione un valor entre 1-5.\n")

# Se ejecuta el programa
menu()
```
